# A Lifecycle Economic Approach to Guiding Savings and Portfolio Decisions


## Introduction

This notebook introduces a **lifecycle model approach** to retirement planning
and portfolio management to quantitative analysts and financial modellers
working in the financial services industry.
By solving and generating scenarios from a lifecycle consumption-saving model, 
we illustrate how structural economic models can provide personalized, transparent,
and data-driven financial advice for savings, retirement decumulation and 
portfolio management.

As **life expectancies increase** and **retirement horizons extend**, 
the financial services industry faces growing pressure to deliver strategies 
that are both **rigorous** and **practically sound**. While many financial 
institutions and advisory firms claim to offer lifecycle-based recommendations, 
these often rely on **ad hoc heuristics** or opaque decision rules that lack 
clear theoretical or mathematical foundations. Such approaches often fail to 
address the **individualized situations and preferences** of clients, 
particularly differences in **wealth, income, income uncertainty,** 
and **preferences related to risk**. 

Recent advancements in computational tools and economic theory enable the 
development of **flexible, estimated structural models**. These models explicitly
and mathematically account for key drivers that shape financial behavior over 
the lifecycle, including:

- **Precautionary saving** to mitigate income risk poses to consumption during
     working years.
- **Accumulating assets for retirement** during peak earning years and as one
     approaches retirement.
- **Decumulating wealth in retirement**, with considerations for bequest motives
     and portfolio risk.

These drivers directly influence key decisions such as:
- **How much to save** during working years to maintain financial stability.
- **What level of portfolio risk to assume** given individual risk tolerance 
    and investment goals.
- **How to allocate wealth over time**, balancing consumption, saving, and
     potential bequest goals.

The key advantage of using a structural approach is that each of the drivers 
which influence how individuals should optimally behave can be pinned down 
by clear parameters -- the model is then solved based on these parameters, allowing
individuals to clearly see how their preferences and characteristics translate
into a personalized financial plan.

## Lifecycle Model: Consumption, Saving, and Portfolio Choice

We now briefly describe the lifecycle model we will use
to generate the scenarios in this notebook. For a comprehensive step by
step mathematical overview of the model, please refer to the [LifeCycle
Prime Time working paper](https://github.com/econ-ark/Life-Cycle-Prime-Time/blob/4417d7a558f3ddaa927cc3d2403f69e0221a44bb/docs/life-cycle-prime-time.pdf).

If you are new to the HARK toolkit, we recommend starting with the
"Gentle Introduction to HARK" notebook.

A **lifecycle model** describes how an individual optimally allocates
resources (spending vs. saving) over time, factoring in income dynamics
and the uncertainties that arise at different life stages. We begin with
the **basic consumption–saving framework**, where each period $t$:

- The agent has **market resources** $\mLvl_t$, consisting of prior net
  worth plus current income $\yLvl_t$.
- They choose how much to **consume**, $\cLvl_t$, subject to a
  **no-borrowing constraint** ($\aLvl_t \ge 0$).
- Remaining **assets** $\aLvl_t = \mLvl_t - \cLvl_t$ earn a return and
  become next period’s bank balance.

#### CRRA Utility and Income Shocks

We assume **Constant Relative Risk Aversion (CRRA)** preferences:
$$
u(\cLvl) \;=\; \frac{\cLvl^{\,1-\CRRA}}{\,1-\CRRA\,},
$$
so the consumer gains diminishing additional utility from each extra
unit of consumption.

Crucially, **uncertainty** enters through **permanent** and
**transitory** shocks to income, following a simplified version of
Friedman’s (1957) setup:

- **Permanent income** $\pLvl_t$ evolves with expected growth plus random
  shocks $\psi_{t+1}$.
- **Transitory income** $\tranShk_{t+1}$ covers events like unemployment
  spells or bonuses, causing year-to-year income variation.

Optimal decisions thus balance **current consumption** against the
desire to **preserve assets** for uncertain future needs and to save for
retirement when income falls. Mathematically,
the solution uses **dynamic programming**: the value of market resources
$\mLvl_t$ today depends on the best possible outcome of future consumption
choices, weighted by survival probabilities and the intertemporal discount factor.

#### Extending to Portfolio Choice a Bequest Motive

After laying out the basic lifecycle saving problem, we introduce an
option to invest in:
- A **risk-free asset** with return factor $\Rfree$.
- A **risky asset** whose returns follow a lognormal process, typically
  with a **positive equity premium** $\eprem$.

In each period, the consumer now chooses **both** the consumption level
$\cLvl_t$ **and** the fraction $\varsigma_t$ of remaining assets
$\aLvl_t$ allocated to the risky asset. This extended **Life Cycle
Portfolio (LCP)** model captures how people diversify between safe and
risky assets to manage uncertainty.

The **basic LCP model** assumes that all wealth is ultimately spent by
the end of life. However, many households exhibit **slower drawdowns**
of wealth than the model predicts, suggesting an additional **bequest
motive**: individuals may derive utility from leaving assets to heirs or
simply enjoy holding wealth (“warm glow”). Mathematically, this motive is
introduced by adding a **bequest utility term** to the agent’s objective.
While some surveys (e.g., SCF) show explicit “leave an estate” motives to
be relatively uncommon, the bequest channel remains as one of the
mechanisms for explaining why retirees often maintain wealth well into
advanced ages.

<div style="

In what follows, we detail the model normalized by permanent income.

At each period $t$, the household seeks to maximize its lifetime utility,
represented by the Bellman equation:

$$
\begin{aligned}
v_t(m_t,S_t) = \max_{c_t,S^{*}_t} \Bigg\{\, &u(c_t) +
\DiePrb_{t+1} \, u_{Beq}(a_t) \\
&+ \DiscFac (1-\DiePrb_{t+1}) 
\Ex_{t} \Big[(\PermGroFac_{t+1}\psi_{t+1})^{1-\CRRA} \\
&\qquad\qquad\qquad\qquad\qquad\times 
v_{t+1}(m_{t+1},S_{t+1}) \Big] \Bigg\}.
\end{aligned}
$$

- $m_t$: Current normalized cash-on-hand or resources.
- $c_t$: Consumption chosen at time $t$.
- $S^*_t$: The risk portfolio share chosen at time $t$ as the end of
  period portfolio choice.
- $S_{t+1}$: The risky portfolio share taken into time $t+1$.

The household gains immediate utility from consumption $u(c_t)$. In the
case of death (with probability $\DiePrb_{t+1}$), it derives utility from
leaving a bequest $u_{Beq}(a_t)$. If the household survives (with
probability $1-\DiePrb_{t+1}$), it discounts future value by $\DiscFac$
and considers the expected continuation value, adjusting for permanent
income growth $\PermGroFac_{t+1}$ and shock $\psi_{t+1}$.

The optimization is subject to several constraints:

1. **Budget Constraint and Savings:**

   End of period normalized assets must balance the household budget
   constraint as follows:
   $$
   a_t = m_t - c_t,
   $$
   where $a_t$ denotes assets held after consumption (end of period
   assets).

   A borrowing constraint ensures:
   $$
   a_t \geq \underline{a}.
   $$

2. **Evolution of Resources:**

   Normalized market resources at the beginning of a period are given by
   the rate of return applied to assets brought over from the previous
   period plus any shocks to income or endowments:
   $$
   m_{t+1} = 
   \frac{\mathsf{R}_{t+1}}{\PermGroFac_{t+1} \psi_{t+1}} \, a_t 
   + \theta_{t+1},
   $$
   where:
   - $\mathsf{R}_{t+1}$: Stochastic return factor.
   - $\theta_{t+1}$: Stochastic income or endowment shock.
   - $\PermGroFac_{t+1}$ and $\psi_{t+1}$: Permanent and transitory
     income shocks respectively.

3. **Rate of Return:**

   The rate of return on a portfolio is given by the risky and safe
   rate of return applied to the assets held in the risky and safe
   portfolio:
   $$
   \mathsf{R}_{t+1} = 
   S_{t+1}\phi_{t+1}\mathbf{R}_{t+1} 
   + (1-S_{t+1})\mathsf{R}_{t+1},
   $$
   where $\mathbf{R}_{t+1}$ is the stochastic risky rate of return and
   $\mathsf{R}_{t+1}$ is the deterministic safe rate of return.

4. **Risky Portfolio Share Evolution:**

   Households only change their risky portfolio with a probability
   $p_t$, thus the beginning of period risky portfolio share is given by:
   $$
   S_{t+1} = 
   \begin{cases}
   S^{*}_t & \text{if } p_t < \wp,\\
   S_t & \text{if } p_t \geq \wp,
   \end{cases}
   $$

5. **Stochastic Shocks:**
   $$
   (\psi_{t+1},\theta_{t+1},\phi_{t+1},p_t) \sim F_{t+1},
   $$
   meaning that income shocks $\psi_{t+1}, \theta_{t+1}$, productivity shock
   $\phi_{t+1}$, and decision-relevant probability $p_t$ are drawn from a
   known distribution.

The utility functions are:

- **Consumption Utility:**
  $$
  u(c) = \frac{c^{1-\CRRA}}{1-\CRRA},
  $$
  where $\CRRA$ is the coefficient of relative risk aversion, representing
  the household’s risk preferences over uncertain future consumption.

- **Bequest Utility:**
  $$
  u_{Beq} (a) = 
  \textbf{BeqFac} \cdot 
  \frac{(a+\textbf{BeqShift})^{1-\CRRA_{Beq}}}{1-\CRRA_{Beq}},
  $$
  where:
  - $\textbf{BeqFac}$: Bequest motive strength.
  - $\textbf{BeqShift}$: Bequest shift parameter.
  - $\CRRA_{Beq}$: Risk aversion specific to bequests.

We refer readers to the [LifeCycle Prime Time working paper](https://github.com/econ-ark/Life-Cycle-Prime-Time/blob/4417d7a558f3ddaa927cc3d2403f69e0221a44bb/docs/life-cycle-prime-time.pdf)
for further discussion on the model.

#### Decumulation and risk 

Let us start with studying how varying risk preference effects the post-retirement 
an individual may plan for decumulation path. Consider an individual with a 
wealth to permanent income ratio of 6.1, who is about to retire at age 65.
We generate four agent types, with the risk aversion parameter taking on
values lower and higher than the baseline. The steps to generate these scenarios
are the same as for generating the baseline population, except we start the 
initial age at 65 and also save the results for each risk aversion type 
in a dictionary which we later plot. 

> **_NOTE:_**  The parameter levels are compared to baseline. How does one 
calibrate the parameters to match individual stated preferences of risk?

> **_NOTE:_**  AAS to MNW: We need an improved explanation of how we have 
started the agents at a given age in the code. Why is there a -25? Can we

```python
# Initialize dictionaries and parameters
data_dict_assets = {}
data_dict_shares = {}
RiskParams = [9.252342476844415, 1.5, 7, 15]  # Baseline and three alternatives
RiskLabels = ["Baseline", "Low Risk Aversion", "Medium Risk Aversion",\
                "Very High Risk Aversion"]

client_normalized_wealth = 6.1
client_age = 65

# Simulate for each risk parameter
for i in range(len(RiskParams)):
    # Initialize agent with baseline calibration
    portfolio_agent_temp = BequestWarmGlowPortfolioType(**init_calibration)
    
    # Set risk aversion and solve model
    portfolio_agent_temp.CRRA = RiskParams[i]
    portfolio_agent_temp.update()
    portfolio_agent_temp.solve()

    # Track relevant variables:
    portfolio_agent_temp.track_vars = [
        "aNrm", "cNrm", "pLvl", "t_age", "mNrm", "Share", "bNrm"
    ]

    # Set simulation parameters
    # AAS to MNW: 
    # # - Matt, can you please check that below is correct?
    ##  - In particular, we want the age to start at 60, not 25 and 
    ##  - we want the starting normalized wealth to be fixed. 
    portfolio_agent_temp.LivPrb = [1.0] * portfolio_agent.T_cycle

    portfolio_agent_temp.T_sim = portfolio_agent.T_cycle

    portfolio_agent_temp.aNrmInitMean = np.log(client_normalized_wealth)
    
    # Run simulation
    portfolio_agent_temp.initialize_sim()
    portfolio_agent_temp.t_cycle[:] = client_age-25
    portfolio_agent_temp.t_age[:] = client_age-25
    history = portfolio_agent_temp.simulate()

    # Process simulation results
    raw_data = {
        "Age": portfolio_agent_temp.history["t_age"].flatten() + 25,
        "nrmB": portfolio_agent_temp.history["bNrm"].flatten(),
        "nrmC": portfolio_agent_temp.history["cNrm"].flatten(),
        "Share": portfolio_agent_temp.history["Share"].flatten(),
    }

    # Calculate medians by age
    Data = pd.DataFrame(raw_data)
    portfolio_agent_temp.AgeMeans = Data.groupby(["Age"]).median().reset_index()

    # Store results in plotting dictionaries
    data_dict_assets[RiskLabels[i]] = (
        portfolio_agent_temp.AgeMeans.Age,
        portfolio_agent_temp.AgeMeans.nrmB
    )
    data_dict_shares[RiskLabels[i]] = (
        portfolio_agent_temp.AgeMeans.Age,
        portfolio_agent_temp.AgeMeans.Share
    )
    
    print(f"Solved and simulated {RiskLabels[i]}")

# Plot 1: Wealth paths
slick_multiline_plot(
    data_dict=data_dict_assets,
    title="Retirement Wealth Paths by Risk Aversion",
    xlabel="Age",
    ylabel="Wealth to Permanent Income Ratio",
    xlim=(66, 95),
    ylim=(0, None),
    linewidth=2.5,
    legend_loc="upper right",
    save_path="scenario_wealth_paths.png"
)
plt.show()

# Plot 2: Equity shares
multiline_plot(
    data_dict=data_dict_shares,
    title="Equity Share by Risk Aversion",
    xlabel="Age",
    ylabel="Fraction in Risky Asset",
    xlim=(66, 95),
    ylim=(0, 1),
    linewidth=2.5,
    legend_loc="upper right",
    save_path="scenario_equity_shares.png"
)
plt.show()
```
